```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN',
'londON_StockhOlm',
'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})
```

```python
df
```

```python
#split "from_to" column into two different columns
df[['From','To']]=df['From_To'].str.split('_',1,expand=True)
```

```python
#drop the 'From_to' column
df.drop('From_To',axis=1,inplace=True)
```

```python
df
```

```python
#Capitalize the city name
city=[]
for i in df['From']:
    city.append(i.capitalize())
```

```python
to=[]
for i in df['To']:
    to.append(i.capitalize())
```

```python
df['To']=to
```

```python
df["From"]=city
```

```python
df
```

```python
#fill missing value
nul = df['FlightNumber'].isnull()
df['FlightNumber']=nul.groupby((nul.diff() == 1).cumsum()).cumsum()*10 + df['FlightNumber'].ffill()
```

```python
#conver float type to int
df['FlightNumber']=df['FlightNumber'].astype(int)
```

```python
df.head()
```

```python
#delete special characters form tha airline column
import re
df['Airline']=df['Airline'].map(lambda x: re.sub(r'\W+', '', x))
```

```python
df
```

```python
#extract data form a list and make a columns according to data in a list
delays=df['RecentDelays'].apply(pd.Series)
delays.columns=['Delays_1','Delays_2','Delays_3']
df[['Delays_1','Delays_2','Delays_3']]=delays
df.drop('RecentDelays',axis=1,inplace=True)
```
